* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
# Без Numpy
X = np.random.randint(0, 10, (5, 3))
print(X)
product = 1
for i in range(min(len(X), len(X[0]))):
    if X[i][i] != 0:
        product *= X[i][i]
print(product)

# C Numpy
diag = np.diagonal(X)
non_zero = diag[diag != 0]

result = np.prod(non_zero) if non_zero.size > 0 else 0
print(result)
```

* __Задача 2__: Даны два вектора x и y. Проверить, задают ли они одно и то же мультимножество.  
  Например, для x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]) ответ True.

```python
# Без Numpy
x = np.array([1, 2, 2, 4])
y = np.array([4, 2, 1, 2])
print(x)
print(y)
if len(x) != len(y):
    print(False)
else:
    result = sorted(x) == sorted(y)
    print(result)

# С Numpy
result = np.array_equal(np.sort(x), np.sort(y))
print(result)
```

* __Задача 3__: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит ноль. 
 Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ 5.

```python
# Без Numpy
x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])
candidates = []
for i in range(1, len(x)):
    if x[i - 1] == 0:
        candidates.append(x[i])

if candidates:
    print(max(candidates))
else:
    print(None)
# C Numpy
mask = x[:-1] == 0
result = x[1:][mask]

print(result.max() if result.size > 0 else None)
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
# Без Numpy
x = [2, 2, 2, 3, 3, 3, 5]
values = []
counts = []
current = x[0]
count = 1

for i in range(1, len(x)):
    if x[i] == current:
        count += 1
    else:
        values.append(current)
        counts.append(count)
        current = x[i]
        count = 1
values.append(current)
counts.append(count)

print(values, counts)

# C Numpy
import numpy as np

x = np.array([2, 2, 2, 3, 3, 3, 5])

change = np.where(x[1:] != x[:-1])[0] + 1
values = np.concatenate(([x[0]], x[change]))
counts = np.diff(np.concatenate(([0], change, [len(x)])))

print(values, counts)
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
# Без Numpy
import math

X = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

Y = np.array([
    [7, 8],
    [9, 10]
])
dist_matrix = []

for x in X:
    row = []
    for y in Y:
        s = 0
        for i in range(len(x)):
            s += (x[i] - y[i]) ** 2
        row.append(math.sqrt(s))
    dist_matrix.append(row)

for r in dist_matrix:
    print(r)
# C Numpy

diff = X[:, np.newaxis, :] - Y[np.newaxis, :, :]
dist_matrix = np.sqrt(np.sum(diff ** 2, axis=2))

print(dist_matrix)
# C scipy.spatial.distance.cdist
from scipy.spatial.distance import cdist

dist_cdist = cdist(X, Y, metric='euclidean')
print(dist_cdist)
```
